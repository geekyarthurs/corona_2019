```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#Setting SNS
sns.set()
```

```python
df = pd.read_csv("covid_19_data.csv")
```

```python
df.head(3)
```

```python
df.set_index('Last Update', inplace=True)
```

```python
df.index
```

```python
df = df.reindex(pd.to_datetime(df.index))
```

```python
df
```

```python
sample = df.resample('5D').sum()
```

```python
sample
```

```python
sns.lineplot(x='Last Update',y='Deaths',data=sample.reset_index(),label="Deaths")
sns.lineplot(x='Last Update',y='Confirmed',data=sample.reset_index(),label="Confirmed Deaths")

plt.xticks(rotation=45)
```

```python
sns.relplot(x='Last Update',y='Confirmed',data=df.reset_index(),col="Country/Region",label="Confirmed Deaths",kind='line')
plt.xticks(rotation=45)
```
